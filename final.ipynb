## Data pre-processing

```python
import pandas as pd
data = pd.read_csv('BlackFriday.csv')

# 0 for female, 1 for male
data['M'] = pd.get_dummies(data['Gender'], drop_first=True) 

# If both columns == 0: city_category = A
data[['City_B', 'City_C']] = pd.get_dummies(data['City_Category'], drop_first=True)

# If all columns == 0: age = 0-17
ages = pd.get_dummies(data['Age'], drop_first=True, prefix='Age_')
data[ages.columns.values] = ages

# If all columns == 0: occupation = 0
occupations = pd.get_dummies(data['Occupation'], drop_first=True, prefix='Occupation_')
data[occupations.columns.values] = occupations 

# Convert all instances of Stay_In_Current_City_Years == '4+' to 4
# Consider column as numeric instead of strings
data['Stay'] = [4 if s == '4+' else int(s) for s in data['Stay_In_Current_City_Years']]
```
